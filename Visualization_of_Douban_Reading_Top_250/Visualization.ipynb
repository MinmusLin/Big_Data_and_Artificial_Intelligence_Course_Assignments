```python
BookCountByNation = doubanData_new['作者'].value_counts()
BookCountByNation
```

```python
BookCountByNation = doubanData_new.groupby(['国家']).count()['图书名'].sort_values(ascending=False)
BookCountByNation
```

```python
BookCountByNation.index
```

```python
BookCountByNation.values
```

```python
countries = BookCountByNation.index
book_counts = BookCountByNation.values

plt.rcParams['font.sans-serif'] = ['SimHei']

colors = plt.cm.tab20c(np.linspace(0, 1, len(countries)))
fig, ax = plt.subplots(figsize=(15, 15))

wedges, texts, autotexts = plt.pie(book_counts,
                                   labels=countries,
                                   autopct='%1.1f%%',
                                   textprops={'fontsize': 16},
                                   colors=colors,
                                   wedgeprops={'linewidth': 1, 'edgecolor': 'white'},
                                   pctdistance=0.6)

threshold = 4
for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    x = np.cos(np.deg2rad(ang))
    y = np.sin(np.deg2rad(ang))
    horizontalalignment = {-1: 'right', 1: 'left'}[int(np.sign(x))]
    connectionstyle = 'angle,angleA=0,angleB={}'.format(ang)
    kw = dict(arrowprops=dict(arrowstyle="-", connectionstyle=connectionstyle, color='black'), zorder=0, va='center')

    if i >= threshold:
        autotexts[i].set_color('black')
        autotexts[i].set_horizontalalignment('center')
        texts[i].remove()
        autotexts[i].remove()
        plt.annotate(f'{countries[i]} {autotexts[i].get_text()}',
                     xy=(x, y),
                     xytext=(1.1 * np.sign(x), 1.1 * y),
                     horizontalalignment=horizontalalignment,
                     **kw)

plt.legend(wedges, countries, title='国家', loc='center left', bbox_to_anchor=(1.05, 0, 0.5, 1))
plt.title('豆瓣读书Top250各国图书数量分布', fontsize=24)
plt.show()
```
